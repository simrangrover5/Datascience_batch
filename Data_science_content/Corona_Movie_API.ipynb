```python
import requests
import json
from PIL import Image
import matplotlib.pyplot as plt

def movie_data():
    """
        This function is used to return movie data
    """
    
    movie = input("\n Enter movie name : ")
    url = f"http://www.omdbapi.com/?t={movie}&apikey=e22bdd41"
    page = requests.get(url)
    data = json.loads(page.text)
    content = ['Title','Actors','Year','Genre','Released','Director','Plot','Language','imdbRating']
    for i in content:
        print(i,"--->",data[i])
    
    image = data['Poster']
    img = requests.get(image)
    ext = image.split(".")[-1]
    filename = f"{movie}.{ext}"
    with open(filename,'wb') as fp:
        fp.write(img.content)
        fp.close()
    
    img = plt.imread(filename)
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()

    
```

```python
movie_data()
```
