```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(12,5),facecolor="black")
ax[0].plot(x,np.sin(x),'ro',label="Sin(X)")
ax[1].plot(x,np.cos(x),'g--',label="Cos(X)")
ax[2].plot(x,np.tan(x),'b-.',label="Tan(x)")
for i in ax:
    i.set_xticks([])
    i.set_yticks([])
    i.legend()

plt.show()
```

```python
y = np.linspace(-5,5,50)
```

```python
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(y,y**2,color="#123456")
ax2.plot(y,y**3,color="black")
for sp in ["left","right","top","bottom"]:
    for ax in (ax1,ax2):
        if sp in ("left","bottom"):
            ax.spines[sp].set_edgecolor("blue")
        else:
            ax.spines[sp].set_edgecolor("red")
            ax.spines[sp].set_linewidth(2)
ax1.spines['left'].set_edgecolor("blue")
plt.show()
```

```python
print(dir(ax.spines['left']))
```

```python
fig,ax = plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        ax[i][j].plot(y,y**2)
plt.show()
```

```python
fig,ax = plt.subplots(2,2)
ax[0][0].plot(x,np.sin(x),'r*',label="Sin(X)")
ax[0][1].plot(x,np.cos(x),'gh',label="Cos(X)")
ax[1][0].plot(y,y**2,'mo',label="$Y^2$")
ax[1][1].plot(y,y**3,'bx',label="$Y^3$")
for i in range(len(ax)):
    ax[i][0].legend(loc="center")
    ax[i][1].legend(loc="center")
    ax[i][0].set_title("Line Plot",fontdict={"fontsize":25,'color':'#123456'})
    ax[i][1].set_title("Line Plot",fontdict={"fontsize":25,'color':'#123456'})
plt.show()
```
