```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
import json
all_tweets = []
f = open("tweets.txt")
for i in f:
    try:
        all_tweets.append(json.loads(i))
    except:
        pass
for i in all_tweets[:10]:
    print(i)
```

```python
countries = []
for i in all_tweets:
    try:
        countries.append(i["user"]["location"])
    except:
        pass
```

```python
len(countries)
```

```python
countries = pd.Series(countries)
```

```python
countries.value_counts()[:20].plot(kind="bar")
```

```python
countries.value_counts()[:20]
```

```python
def check_value(x):
    if x != None:
        return True
    return False

countries[countries.apply(check_value)].value_counts()
```

```python
hashtags = []
for i in all_tweets:
    try:
        for word in i["text"].split():
            if "#" in word:
                hashtags.append(word)
    except:
        pass
```

```python
hashtags
```

```python
fig = plt.figure(facecolor="#abcdef")
ax = fig.add_axes([0,0,1,1],facecolor="#abcdef")
hashtags = pd.Series(hashtags)
ax.bar(hashtags.value_counts().index[:10],hashtags.value_counts().values)
plt.show()
```

```python
hashtags.value_counts().index
```

```python
newhash = []
for i in hashtags:
    if i[1].lower() in list("abcdefghijklmnopqrstuvwxyz"):
        newhash.append(i)
```

```python
hashes = pd.Series(newhash)
```

```python
hashes.value_counts()
```

```python
country = []
state = []
text = []
lang = []
day = []
columns = ['lang','country','state','text']
for i in all_tweets:
    try:
        if i["lang"].lower() in ["en","hi"]:
            if "," in i["user"]["location"]:
                s,c = i["user"]["location"].split(",")
                state.append(s)
                country.append(c)
                text.append(i["text"])
                if i["lang"].lower() == "en":
                    lang.append("en")
                else:
                    lang.append("hi")
                day.append(i["created_at"].split()[0])
    except:
        pass
```
