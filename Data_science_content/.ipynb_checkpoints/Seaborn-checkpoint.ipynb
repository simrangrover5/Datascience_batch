```python
sns.scatterplot(iris['sepal_length'],iris['petal_length'])
plt.show()
```

```python
c = iris['species'].map({'setosa':0,'virginica':1,'versicolor':2})
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(iris['sepal_length'],iris['petal_length'],hue=c,palette='rainbow')
plt.show()
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
sns.scatterplot(iris['sepal_length'],iris['petal_length'],hue=c,palette=cmap)
plt.show()
```

```python
sns.scatterplot('total_bill','tip',data=tip,hue="sex",size="smoker")
plt.show()
```

```python
sns.scatterplot('total_bill','tip',data=tip,hue="day")
plt.show()
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
mpg['cylinders'].value_counts()
```

```python
data = mpg[:100]
```

```python
sns.scatterplot('mpg','horsepower',data=data,size='cylinders',color='red')
plt.show()
```

```python
c = data['origin'].map({'usa':0,'europe':1,'japan':2})
```

```python
sns.scatterplot('mpg','horsepower',data=data,size='cylinders',color='red',hue=c,palette='hot')
plt.show()
```
