```python
x = 250
y = 500
img = np.zeros(375000).reshape(250,500,3)
for i in range(x):  #for particular 2d array
    for j in range(y): #for each rows
        img[i][j] = [0,0,0]  #for black
plt.imshow(img)
```

```python
x = 250
y = 500
img = np.zeros(375000).reshape(250,500,3)
for i in range(x):  #for particular 2d array
    for j in range(y): #for each rows
        img[i][j] = [255,87,51]  #for black
plt.imshow(img)
```

```python
x = 250
y = 500
img = np.zeros(375000).reshape(250,500,3)
for i in range(x):  #for particular 2d array
    for j in range(y): #for each rows
        img[i][j] = [255,165,0]  #for black
plt.imshow(img)
```

```python
img[::5]
```

```python
###flag --> height, width
x = 250
y = 500
img = np.zeros(250*500*3).reshape(250,500,3)
for i in range(x):
    for j in range(y):
        if i in range(x//3):
            img[i][j] = [255,165,0]
        
plt.imshow(img)
```
