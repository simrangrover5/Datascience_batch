```python
x = np.linspace(0,2*np.pi,15)
sin = np.sin(x)
```

```python
plt.rcParams["figure.figsize"] = (11,6)
```

```python
label = {
    "fontsize" : 20,
    "color" : "#1234ff",
    "fontstyle" : "italic",
    "fontfamily" : "Sans Serif"
}
title = {
    "fontsize" : 25,
    "color" : "red",
    "fontstyle" : "italic",
    "fontfamily" : "Sans Serif"
}
```

```python
plt.figure(figsize=(11,5))
plt.plot(x,sin,'ro--',markersize=10,markeredgecolor="#123456",markerfacecolor="white")
plt.plot(x,np.cos(x),'g*:',markersize=10,markeredgecolor="#123456",markerfacecolor="white")
plt.xticks(range(0,8),fontsize=15)
plt.yticks(fontsize=15)
plt.title("Sin(X)-Cos(X)",fontdict=title)
plt.xlabel("X",fontdict=label)
plt.ylabel("Sin(X),Cos(x)",fontdict=label)
plt.tick_params(labelcolor="#FF5733",color="#FF5733",size=12,width=5,length=5)
plt.show()
```

```python
print(dir(plt))
```

```python
plt.figure(figsize=(11,5))
plt.plot(x,sin,'ro--',markersize=10,markeredgecolor="#123456",markerfacecolor="white",label="Sin(X)")
plt.plot(x,np.cos(x),'g*:',markersize=10,markeredgecolor="#123456",markerfacecolor="white",label="Cos(X)")
plt.xticks(range(0,8),fontsize=15)
plt.yticks(fontsize=15)
plt.title("Sin(X)-Cos(X)",fontdict=title)
plt.xlabel("X",fontdict=label)
plt.ylabel("Sin(X),Cos(x)",fontdict=label)
plt.tick_params(labelcolor="#FF5733",color="#FF5733",size=12,width=5,length=5)
plt.legend(loc="center")  #loc = 1,2,3,4
plt.show()
```

```python
plt.figure(figsize=(11,5))
plt.plot(x,sin,'ro--',markersize=10,markeredgecolor="#123456",markerfacecolor="white",label="Sin(X)")
plt.plot(x,np.cos(x),'g*:',markersize=10,markeredgecolor="#123456",markerfacecolor="white",label="Cos(X)")
plt.xticks(range(0,8),fontsize=15)
plt.yticks(fontsize=15)
plt.title("Sin(X)-Cos(X)",fontdict=title)
plt.xlabel("X",fontdict=label)
plt.ylabel("Sin(X),Cos(x)",fontdict=label)
plt.tick_params(labelcolor="#FF5733",color="#FF5733",size=12,width=5,length=5)
plt.legend(loc=[0.8,0.7],fontsize=13)  #loc = 1,2,3,4
plt.grid(color="gray")
plt.show()
```

```python
grras_jaipur = np.random.randint(20,25,10)
grras_delhi = np.random.randint(20,25,10)
```
