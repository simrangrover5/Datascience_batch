```python
values = data['STATES'].value_counts()[:10].values
index = data['STATES'].value_counts()[:10].index
fig = plt.figure(figsize=(11,5))
ax = fig.add_axes([0,0,1,1],facecolor="black")
sns.barplot(index,values,ax=ax)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90,color="red")
ax.set_yticklabels(ax.get_yticks(),color="#123456")
ax.text(5,300000,"State Wise Petitions",color="white",fontsize=25)
x,y = ax.patches[0].get_xy()
for i in range(len(index)):
    x = ax.patches[i].get_x()
    y = ax.patches[i].get_height()
    ax.text(x,y,values[i],fontsize=15,color="white")
ax.set_xlabel("States",fontsize=25)
ax.set_ylabel("Count",fontsize=25)
for i in ["left","right","top","bottom"]:
    ax.spines[i].set_color("yellow")
    ax.spines[i].set_linewidth(3)
plt.show()
```

```python
data[data['STATES'] == " TEXAS"]["YEAR"].value_counts()
```

```python
for i in states:
    values = data[data['STATES'] == i]["YEAR"].value_counts().values
    index = data[data['STATES'] == " TEXAS"]["YEAR"].value_counts().index
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1],facecolor="black")
    sns.barplot(index,values,ax=ax,palette="Purples")
    ax.set_title(i,color="red",fontsize=30)
    ax.grid(color="#bbbbbb",linestyle="--",lw=0.5,alpha=0.5)
    ax.set_xlabel("YEAR",fontsize=15)
    ax.set_ylabel("COUNT",fontsize=15)
    plt.show()
```
