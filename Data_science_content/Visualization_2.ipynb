```python
df['species'] = df['species'].map({'setosa':0,'virginica':1,'versicolor':2})
```

```python
df
```

```python
plt.scatter(df['sepal_length'],df['petal_length'],c=df['species'])
plt.xlabel("Sepal Length",fontdict={"color":"red","fontsize":20})
plt.ylabel("Petal Length",fontdict={"color":"red","fontsize":20})
plt.show()
```

```python
plt.scatter(df['sepal_length'],df['petal_length'],c=df['species'],s=100)
plt.xlabel("Sepal Length",fontdict={"color":"red","fontsize":20})
plt.ylabel("Petal Length",fontdict={"color":"red","fontsize":20})
plt.show()
```

```python
plt.hist(df["sepal_width"],bins=30,ec='k')
plt.show()
```
