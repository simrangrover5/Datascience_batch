```python
def create_dfs(soup_object):
    try:
        actual_price = [i.text for i in soup_object.select("._3auQ3N._2GcJzG")]
        dis_price = [i.text for i in soup_object.select("._1vC4OE._2rQ-NK")]
        names = []
        ratings = []
        for i in soup_object.select(".col.col-7-12"):
            names.append(i.select("._3wU53n")[0].text)
            if i.select(".hGSR34"):
                ratings.append(i.select(".hGSR34")[0].text)
            else:
                ratings.append(None)
        print(actual_price,len(actual_price))
        print(dis_price,len(dis_price))
        print(names,len(names))
        print(ratings,len(ratings))
        if len(names) == len(ratings) == len(actual_price) == len(dis_price):
            df = pd.DataFrame({
                "name" : names,
                "ratings" : ratings,
                "Actual Price": actual_price,
            "Discount_price" : dis_price
            })
            return df
        else:
            return "The length of all the data is not same"
    except Exception as e:
        print("Error : ",e)
```
