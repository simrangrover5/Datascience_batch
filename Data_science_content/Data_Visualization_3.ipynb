```python
x = np.random.randint(1,10,20)
y1 = 2*x**2 + 3*x + 3
```

```python
fig = plt.figure(figsize=(11,6))
plt.plot(x,y1,'o--')
```

```python
fig = plt.figure(figsize=(11,6))  #facecolor="#123456"
ax1 = fig.add_axes([0,0,1,1],facecolor="black")  #[x,y,height,width] 1,1 --> 100%, 100%
ax1.plot(x,y1,'ro--')
ax1.set_title("Polynomial Graph",color="#123456",fontdict={"fontsize":30,"fontstyle":"italic"})
ax1.set_xlabel("X",fontdict={"color":'brown','fontsize':20,"fontfamily":"monospace"})
ax1.set_ylabel("Y",fontdict={"color":'brown','fontsize':20,"fontfamily":"monospace"})
ax1.set_xticks(range(0,11))
ax1.set_xticklabels(range(0,11),fontsize=15,color="blue")  #get_xticks
ax1.set_yticklabels(range(0,220),fontsize=15,color="blue")
plt.show()
```

```python
print(dir(ax1))
```

```python
fig = plt.figure(figsize=(11,6))  #facecolor="#123456"
ax1 = fig.add_axes([0,0,1,1],facecolor="black")#[x,y,height,width] 1,1 --> 100%, 100%
ax1.plot(x,y1,'ro--')
ax1.set_title("Polynomial Graph",color="#123456",fontdict={"fontsize":30,"fontstyle":"italic"})
ax1.set_xlabel("X",fontdict={"color":'brown','fontsize':20,"fontfamily":"monospace"})
ax1.set_ylabel("Y",fontdict={"color":'brown','fontsize':20,"fontfamily":"monospace"})
ax1.set_xticks(range(0,11))
ax1.set_xticklabels(ax1.get_xticks(),fontsize=15,color="blue")  #get_xticks
ax1.set_yticklabels(range(0,220),fontsize=15,color="blue")
#ax1.spines["left"].set_visible(False)
for sp,c in zip(["left","right","top","bottom"],["green","green","yellow","magenta"]):
    ax1.spines[sp].set_edgecolor(c)
    ax1.spines[sp].set_linewidth(4)
plt.show()
```

```python
x = np.linspace(-5,5,20)
```

```python
fig = plt.figure(figsize=(11,6),facecolor="black")
#ax1 = fig.add_axes([0,0,1,1])
#ax1.plot(x,x)
ax2 = fig.add_axes([0,0,0.5,0.5],facecolor="#123456")
ax2.plot(x,x**2,'r',label="$X^2$",lw=2,alpha=0.8)
ax3 = fig.add_axes([0.5,0.5,0.5,0.5],facecolor="#abcdef")
ax3.plot(x,x**3,'g',label="$X^3$",lw=2,alpha=0.5)
ax2.legend()
ax3.legend()
for sp in ["left","right","top","bottom"]:
    for i in [ax2,ax3]:
        i.spines[sp].set_visible(False)
ax2.set_xticklabels(ax2.get_xticks(),color="white")
ax2.set_yticklabels(ax2.get_yticks(),color="white")
ax3.set_xticklabels(ax3.get_xticks(),color="white")
ax3.set_yticklabels(ax3.get_yticks(),color="white")
plt.show()
```

```python
fig = plt.figure(figsize=(11,6),facecolor="black")
ax1 = fig.add_axes([0,0.5,0.5,0.5],facecolor="black")
ax1.plot(x,x,'--',label="X")
ax2 = fig.add_axes([0,0,0.6,0.5],facecolor="black")
ax2.plot(x,x**2,'r:',label="$X^2$",lw=2,alpha=0.8)
ax3 = fig.add_axes([0.5,0.5,0.5,0.5],facecolor="black")
ax3.plot(x,x**3,'g-.',label="$X^3$",lw=2,alpha=0.8)
ax2.legend()
ax3.legend()
ax1.legend()
ax4 = fig.add_axes([0.6,0,0.5,0.5],facecolor="black")
ax4.plot(x,x**4,'m--',label="$X^4$")
ax4.legend()
for sp in ["left","right","top","bottom"]:
    for i in [ax2,ax3,ax1]:
        i.spines[sp].set_visible(False)
        i.set_xticks([])
        i.set_yticks([])
ax2.set_xticklabels(ax2.get_xticks(),color="white")
ax2.set_yticklabels(ax2.get_yticks(),color="white")
ax3.set_xticklabels(ax3.get_xticks(),color="white")
ax3.set_yticklabels(ax3.get_yticks(),color="white")
plt.show()
```
