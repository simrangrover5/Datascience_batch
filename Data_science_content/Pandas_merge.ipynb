### To find the outliers there is two approach

    1. z_score
    2. IQR
    
    
    z_score = observation-mean/std  , compare it with 1
    
    IQR --> Inter Quartile Range
            (Q3 - Q1)/2
            
            lower_bound = Q1 - (IQR * 1.5)
            upper_bound = Q3 + (IQR * 1.5)
            
            The observations below the lower bound and above the upper bound are outliers

```python
arr = np.random.randint(10,50,100)
arr =np.append(arr,np.random.randint(200,300,50))
np.random.shuffle(arr)
```

```python
arr
```

```python
df = pd.DataFrame(arr.reshape(30,5),columns = ['A',"B","C","D","E"])
```

```python
df
```

```python
for i in df:
    plt.hist(df[i],ec='k')
    plt.show()
```

```python
outlier = {}
for i in df:  #for column
    o = []
    for j in df[i]:
        mean = df[i].mean()
        std = df[i].std()
        z_score = (j - mean)/std
        if z_score > 0:
            o.append(j)
    outlier[i] = o
    
print(outlier)
```
