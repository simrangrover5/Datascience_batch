```python
from time import sleep
x = list(range(50))
y = [psutil.cpu_percent() for i in range(50)]
fig = plt.figure(facecolor="black")
ax = fig.add_axes([0,0,1,1],facecolor="black")
ax.plot(x,y,'ro--')
def change_plot(wait):
    global y
    #sleep(wait)
    ax.clear()
    y.pop(0)
    y.append(psutil.cpu_percent())
    plt.plot(x,y,'ro--')
    

func = FuncAnimation(fig,change_plot,2)
plt.show()
```
