```python
from random import randint
from time import sleep
y = list(np.random.randint(10,20,20))
dates = pd.date_range("20200106",periods=20)
fig,ax = plt.subplots()
ax.plot(dates,y,'ro:')
#ax.set_xticks(rotation=30)
def change_plot(wait=None):
        global y
        #sleep(wait)
        ax.clear()
        y.pop(0)
        y.append(randint(10,20))
        ax.plot(dates,y,'ro:')
        
func = FuncAnimation(fig,change_plot)
plt.show()
```
